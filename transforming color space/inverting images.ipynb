Sometimes the inverted image in the lab space looks more natural than the image in the RGB space.

```python
import numpy as np
from skimage.io import imread
from skimage.color import rgb2lab, lab2rgb
import matplotlib.pylab as plt
```

```python
im = imread('dataset/transforming color space/Surface_Laptop_3_04.jpg')
im1 = rgb2lab(im)
# inverting the image
im1 = rgb2lab(im)
im1[...,0] = np.max(im1[...,0]) - im1[...,0]
im1 = lab2rgb(im1)
plt.figure(figsize=(20,10))
plt.subplot(121), plt.imshow(im), plt.axis('off'),
plt.title('Original image', size=20)
plt.subplot(122), plt.imshow(im1), plt.axis('off'),
plt.title('Gray scale image', size=20)
plt.show()
```

```python
im = imread('dataset/transforming color space/Surface_Laptop_3_04.jpg')
# inverting the image

plt.figure(figsize=(20,10))
plt.subplot(121), plt.imshow(im), plt.axis('off'),
plt.title('Original image', size=20)
plt.subplot(122), plt.imshow(im1), plt.axis('off'),
plt.title('Gray scale image', size=20)
plt.show()
```

```python
# Importing imagechops for using the invert() method
from PIL import Image, ImageChops
```

```python
# Opening the test image, and saving it's object
img = Image.open('dataset/transforming color space/Surface_Laptop_3_04.jpg')
 
# Passing the image object to invert()  
inv_img = ImageChops.invert(img)
 
```

```python
plt.figure(figsize=(20,10))
plt.subplot(121), plt.imshow(im), plt.axis('off'),
plt.title('Original image', size=20)
plt.subplot(122), plt.imshow(inv_img), plt.axis('off'),
plt.title('Inverted image(RGB)', size=20)
plt.show()
```
